```python
import numpy as np
import pandas as pd
import plotly.express as px
import os

# from pandasql import sqldf
```

```python
baseFileLocation = 'G:/Survey_Live/20260401/'
fileExtension = '.csv'
fileTypes=['Values', 'Labels']
fileStatuses = ['complete', 'incomplete']
groups = [1,2]
fileReadFlag = True
fileNameList = []
```

```python
def checkLocation(location: str):
    if not os.path.exists(location):
        os.makedirs(name=location, exist_ok=True)
```

```python
# filepath = Path("folder/subfolder/my_file.csv")

# # Create the parent directory (and any necessary intermediate parents)
# # exist_ok=True prevents an error if the folder already exists
# filepath.parent.mkdir(parents=True, exist_ok=True)

fileReadFlag = True
survey_data = pd.DataFrame()

for fileType in fileTypes:
    
    sourceFileLocationName = baseFileLocation + 'Visual_FactSheet_Live_' + fileType + fileExtension
    
    fileLocation = baseFileLocation + fileType + '/'
    saveFileLocation =''
    
    if fileReadFlag:
        survey_data = pd.read_csv(sourceFileLocationName, skiprows=[1,2])
        survey_data = survey_data[(survey_data['DistributionChannel'] == 'anonymous') | (survey_data['DistributionChannel'] == 'qr')]
        
        fileReadFlag = False

    for fileStatus in fileStatuses:
        if fileStatus == 'complete':
            competeSurvey = survey_data[survey_data['Progress'] == 100]
            saveFileLocation = fileLocation + fileStatus + '/'
            saveFileLocationName= saveFileLocation + fileStatus + '_survey_' + fileType + fileExtension
            checkLocation(saveFileLocation)
            competeSurvey.to_csv(saveFileLocationName, index=False)

            for group in groups:
                groupFileLocationName= saveFileLocation + fileStatus + '_group_' + str(group) + '_data_' + fileType + fileExtension
                checkLocation(saveFileLocation)
                fileNameList.append(groupFileLocationName)
                group_data = competeSurvey[competeSurvey['Group'] == group]
                group_data.to_csv(groupFileLocationName, index=False)
        
        else:
            incompeteSurvey = survey_data[survey_data['Progress'] != 100]
            saveFileLocation= fileLocation + fileStatus + '/'
            saveFileLocationName= saveFileLocation + fileStatus + '_survey_' + fileType + fileExtension
            checkLocation(saveFileLocation)
            incompeteSurvey.to_csv(saveFileLocationName, index=False)

            for group in groups:
                groupFileLocationName= saveFileLocation + fileStatus + '_group_' + str(group) + '_data_' + fileType + fileExtension
                checkLocation(saveFileLocation)
                fileNameList.append(groupFileLocationName)
                group_data = incompeteSurvey[incompeteSurvey['Group'] == group]
                group_data.to_csv(groupFileLocationName, index=False)


fileReadFlag = True
# del survey_data
```

```python
questionList = [3,4,7,8,9,10,11,14,17,18,19,20,21,23]

for inputFile in fileNameList:
    location = inputFile

    competeSurey = pd.read_csv(location)

    for i in range(1,28):
        if i not in questionList:
            continue

        copy = competeSurey.copy()
        question = 'Q' + str(i)
        for col in competeSurey.columns:
            if (col.startswith('Q')):
                if not col.startswith(question):
                    copy.drop(col, axis=1, inplace=True)
        
        fileLocation = 'pilot_results/data_by_question/' + location.split('/')[1] + '/' + location.split('/')[2]
        fileName = location.split('/')[3].split('.')[0]
        checkLocation(fileLocation)
        copy.to_csv(fileLocation + '/' + question + '_' + fileName + fileExtension, index=False)
        del copy

print('Done')
```
